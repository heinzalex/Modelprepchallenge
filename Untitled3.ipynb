```python
life_df.columns = life_df.columns.str.strip()
```

```python
#missing values and outliers
life_df.isnull().sum()*100/life_df.isnull().count()
```

```python
#lstrip, strip(), rstrip

life2_df = life_df.copy()

fill_list = ['Life expectancy', 'Adult Mortality', 'Alcohol', 
             'Hepatitis B','BMI', 'Polio', 'Total expenditure', 'Diphtheria', 'GDP', 'Population',
             'thinness  1-19 years', 'thinness 5-9 years', 'Income composition of resources', 
            'Schooling']

countries = life2_df['Country'].unique()

for col in fill_list:
    for country in countries:
        life2_df.loc[life2_df['Country'] == country, col] = life2_df.loc[
life2_df['Country'] == country, col].fillna(life2_df[life2_df['Country'] == country]
[col].mean())
#Compare with country
```

```python
life3_df = life_df.copy()

for col in fill_list:
    life3_df.loc[:, col] = life3_df.loc[:, col]. interpolate(limit_direction='both')

life3_df.dropna(inplace=True)
```

```python
life3_df['winsorized_pop'] = winsorize(life3_df["Population"], (0, 0.05))

life3_df['winsorized_gdp'] = winsorize(life3_df['GDP'], (0, 0.05))

list(life3_df)
```
